```python
pop09 = pd.read_csv("insee/population-legales-2009.csv", sep=";", dtype={'CODDEP':str})
pop09 = pop09[['CODDEP','PMUN']]
pop09['PMUN'] = pop09['PMUN'].apply(lambda x: str(x).replace(" ",""))
pop09['PMUN'] = pop09['PMUN'].astype('int64')
pop09 = pop09.rename(columns={'CODDEP':'C_DEPT','PMUN':'POPULATION'})
```

```python
pop11 = pd.read_csv("insee/population-legales-2011.csv", sep=";", dtype={'CODDEP':str})
pop11 = pop11[['CODDEP','PMUN']]
pop11['PMUN'] = pop11['PMUN'].apply(lambda x: str(x).replace(" ",""))
pop11['PMUN'] = pop11['PMUN'].astype('int64')
pop11 = pop11.rename(columns={'CODDEP':'C_DEPT','PMUN':'POPULATION'})
```

```python
pop13 = pd.read_csv("insee/population-legales-2013.csv", sep=";", dtype={'CODDEP':str})
pop13 = pop13[['CODDEP','PMUN']]
pop13['PMUN'] = pop13['PMUN'].apply(lambda x: str(x).replace(" ",""))
pop13['PMUN'] = pop13['PMUN'].astype('int64')
pop13 = pop13.rename(columns={'CODDEP':'C_DEPT','PMUN':'POPULATION'})
```

```python
pop15 = pd.read_csv("insee/population-legales-2015.csv", sep=";", dtype={'CODDEP':str})
pop15 = pop15[['CODDEP','PMUN']]
pop15['PMUN'] = pop15['PMUN'].apply(lambda x: str(x).replace(" ",""))
pop15['PMUN'] = pop15['PMUN'].astype('int64')
pop15 = pop15.rename(columns={'CODDEP':'C_DEPT','PMUN':'POPULATION'})
```

```python
pop17 = pd.read_csv("insee/population-legales-2017.csv", sep=";", dtype={'CODDEP':str})
pop17 = pop17[['CODDEP','PMUN']]
pop17['PMUN'] = pop17['PMUN'].apply(lambda x: str(x).replace(" ",""))
pop17['PMUN'] = pop17['PMUN'].astype('int64')
pop17 = pop17.rename(columns={'CODDEP':'C_DEPT','PMUN':'POPULATION'})
```

```python
dfsinoe2009pop = pd.merge(dfsinoe2009,pop09,on='C_DEPT',how='left')
dfsinoe2011pop = pd.merge(dfsinoe2011,pop11,on='C_DEPT',how='left')
dfsinoe2013pop = pd.merge(dfsinoe2013,pop13,on='C_DEPT',how='left')
dfsinoe2015pop = pd.merge(dfsinoe2015,pop15,on='C_DEPT',how='left')
dfsinoe2017pop = pd.merge(dfsinoe2017,pop17,on='C_DEPT',how='left')
```

```python
dfsinoe2009pop['RATIO'] = (dfsinoe2009pop['TONNAGE'] * 1000) / dfsinoe2009pop['POPULATION']
dfsinoe2011pop['RATIO'] = (dfsinoe2011pop['TONNAGE'] * 1000) / dfsinoe2011pop['POPULATION']
dfsinoe2013pop['RATIO'] = (dfsinoe2013pop['TONNAGE'] * 1000) / dfsinoe2013pop['POPULATION']
dfsinoe2015pop['RATIO'] = (dfsinoe2015pop['TONNAGE'] * 1000) / dfsinoe2015pop['POPULATION']
dfsinoe2017pop['RATIO'] = (dfsinoe2017pop['TONNAGE'] * 1000) / dfsinoe2017pop['POPULATION']
```

```python
dfsinoe2017pop = dfsinoe2017pop[dfsinoe2017pop['C_TYP_REG_DECHET'] != '02C']
dfsinoe2017pop = dfsinoe2017pop[dfsinoe2017pop['C_TYP_REG_DECHET'] != '07A']
dfsinoe2017pop = dfsinoe2017pop[dfsinoe2017pop['C_TYP_REG_DECHET'] != '07B']
dfsinoe2017pop = dfsinoe2017pop[dfsinoe2017pop['C_TYP_REG_DECHET'] != '07C']
```

### Par type nat

```python
tonnagepertype = dfsinoe2017pop[['C_TYP_REG_DECHET','TONNAGE']].groupby("C_TYP_REG_DECHET", as_index=False).sum()
```

```python
tonnagepertype['RATIO'] = (tonnagepertype['TONNAGE'] * 1000) / popperdep.POPULATION.sum()
```

```python
tonnagepertype = pd.merge(tonnagepertype, libelletypedechet, on='C_TYP_REG_DECHET', how='left')
```

```python
tonnagepertype
```

```python
jsonnat2017 = []
```

```python
dictnat2017 = {}
```

```python
dictnat2017['tonnage'] = tonnageperdep.TONNAGE.sum()
dictnat2017['kg_par_habitant'] = tonnageperdep.TONNAGE.sum() * 1000 / popperdep.POPULATION.sum()
dictnat2017['infos_per_type'] = []
for index, row in tonnagepertype.iterrows():
    dictpertype = {}
    dictpertype['code_type'] = row['C_TYP_REG_DECHET']
    dictpertype['tonnage'] = row['TONNAGE']
    dictpertype['kg_par_habitant'] = row['RATIO']
    dictpertype['libelle_type'] = row['L_TYP_REG_DECHET']
    
    dictnat2017['infos_per_type'].append(dictpertype)
```

```python
jsonnat2017.append(dictnat2017)
```

```python
jsonnat2017
```

```python
import json

with open('dechets-maille-national.json', 'w') as fp:
    json.dump(jsonnat2017, fp)
```

```python
for ind,roww in dfreg.iterrows():
    dictonereg2017 = {}
    
    dictonereg2017['tonnage'] = roww['TONNAGE']
    dictonereg2017['kg_par_habitant'] = roww['RATIO']
    dictonereg2017['libelle'] = roww['L_REGION']
    dictonereg2017['reg'] = roww['C_REGION']
    dictonereg2017['infos_per_type'] = []
    for index, row in tonnagepertypereg.iterrows():
        if(roww['C_REGION'] == row['C_REGION']):
            dictpertypereg = {}
            dictpertypereg['code_type'] = row['C_TYP_REG_DECHET']
            dictpertypereg['tonnage'] = row['TONNAGE']
            dictpertypereg['kg_par_habitant'] = row['RATIO']
            dictpertypereg['libelle_type'] = row['L_TYP_REG_DECHET']
            dictonereg2017['infos_per_type'].append(dictpertypereg)
    dictreg2017.append(dictonereg2017)
```
